```python
import gensim
```

```python
from gensim.models import Word2Vec,KeyedVectors
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story=[]

for filename in os.listdir("sherlock"):
    f=open(os.path.join("sherlock", filename))
    corpus=f.read()
    raw_sent=sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
len(story)
```

```python
story[0][0:20]
```

```python
smodel=gensim.models.Word2Vec(
                            window=10,
                            min_count=2,
                            workers=4)
```

```python
smodel.build_vocab(story)
```

```python
smodel.train(story,total_examples=smodel.corpus_count, epochs=smodel.epochs)
```

```python
smodel.wv.most_similar('sherlock')
```

smodel.wv.doesnt_match("watson","sherlock","pipe","mycroft")

```python
smodel.wv.doesnt_match(["sherlock","irene","mycroft"])
```

```python
smodel.wv.similarity("sherlock","watson")
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Pick top interesting words from your corpus or model
words = ['sherlock', 'holmes', 'watson', 'murder', 'detective', 'crime', 'baker', 'london', 'case', 'scotland']

# Create similarity matrix
similarity_matrix = np.zeros((len(words), len(words)))

for i, word1 in enumerate(words):
    for j, word2 in enumerate(words):
        similarity_matrix[i][j] = smodel.wv.similarity(word1, word2)

# Convert to DataFrame for seaborn
df_sim = pd.DataFrame(similarity_matrix, index=words, columns=words)

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_sim, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Word2Vec Similarity Heatmap')
plt.show()
```
